```python
a = float(input("enter length of first side :"))
b = float(input("enter length of 2nd side :"))
c = float(input("enter length of 3rd side :"))
s = (a+b+c)/2
Area = (s*(s-a)*(s-b)*(s-c))**0.5
print("this is area of traingle %0.2f" %Area)
```
